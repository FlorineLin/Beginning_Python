要理解 “Tuples: Immutable Sequences”（元组：不可变序列），核心可拆解为两个关键属性 ——“序列” 本质与 **“不可变” 特性 **，具体解析如下：
1. 首先是 “Sequence（序列）”：元组的基础形态
元组是 Python 中典型的 “序列类型”，和列表（List）、字符串（String）共享序列的核心特征：
有序存储：元素按插入顺序排列，可通过 “索引（Index）” 访问单个元素（如 tup[0] 取第一个元素），也支持 “切片（Slicing）” 获取子序列（如 tup[1:3] 取第 2-3 个元素）；
可包含任意元素：能混合存储不同类型的数据，比如整数、字符串、列表甚至另一个元组（如 (1, "hello", [3,4], (5,6))）；
支持序列通用操作：可通过 len(tup) 求长度、x in tup 判断元素是否存在、tup1 + tup2 拼接两个元组等。
2. 关键是 “Immutable（不可变）”：元组与列表的核心区别
“不可变” 是元组最独特的属性，指元组创建后，其元素的 “指向” 无法修改——
不能新增 / 删除元素：比如无法用 tup.append(6)（列表支持）给元组加元素，也不能用 del tup[0] 删除元素；
不能修改元素本身（针对 “不可变元素”）：若元组元素是整数、字符串等不可变类型，直接修改会报错（如 tup = (1,2); tup[0] = 3 会触发 TypeError）；
注意 “嵌套可变元素” 的特殊情况：若元组包含列表等可变类型，列表内部元素可改（如 tup = (1, [2,3]); tup[1][0] = 4 是允许的），但元组对这个列表的 “指向” 仍无法改变（不能把列表换成其他元素）。
总结
元组的本质是 “不可变的序列”：既保留了序列 “有序、可索引、可包含多种元素” 的灵活性，又通过 “不可变” 确保了数据安全性（创建后不会被意外修改），因此常用来存储无需变动的固定数据（如坐标 (x,y)、函数返回的多值组合等）。
